# Predicting the Loan Status of a Customer Using Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model 
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder 


# Importing the dataset 
input_file = "Dataset.csv"
df = pd.read_csv(input_file, header = 0)

# Removing the Rows with empty Values
df=df.dropna()

# Encoding the Categorical Value with Numeric Values 
# (We are doing it to replace the string values with numeric as ML Algorithms only works with numeric values)
# Here I have showed two ways by which you can encode 

# Replacing Categorical Values Using a Dictionary
a={"3+":5,"0":0,"1":1,"2":2}
df['Dependents'] = df['Dependents'].map(a)

# Replacing the Categorical Values Using Label Enoders by importing the package LabelEncoder from sklearn.preprocessing
le = LabelEncoder() 
df['Loan_ID']= le.fit_transform(df['Loan_ID']) 
df['Gender']= le.fit_transform(df['Gender']) 
df['Married']= le.fit_transform(df['Married']) 
df['Education']= le.fit_transform(df['Education']) 
df['Self_Employed']= le.fit_transform(df['Self_Employed']) 
df['Property_Area']= le.fit_transform(df['Property_Area'])
df['Loan_Status']= le.fit_transform(df['Loan_Status']) 
df.head(5)
```

```python
# Defining the Dependent and Independent Variables for our model 
# Note:- We are going to Predict the Dependent Variable

Y = df[['Loan_Status']]   # Dependent Variable 
X = df[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']] # Independent Variable
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30 ,random_state = 0)
```

```python
# Feature Scaling to prevent overfitting

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Logistic Regression to the Training set

classifier = linear_model.LogisticRegression()
m=classifier.fit(X_train, Y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
# Making the Confusion Matrix that will help us predict the Accuracy of the model 

cm = confusion_matrix(Y_test, y_pred)
print(cm) # We Print Our confucion Matrix (Optional)
```

```python
# Using our Confusion Matrix to get the accuracy 

print("Accuracy of Logistic Regression Model is:",
     metrics.accuracy_score(Y_test,y_pred)*100,"%")
X.head()
```
